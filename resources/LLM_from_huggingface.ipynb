```python
import os
from typing import List, Optional
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.llms.huggingface_api import HuggingFaceInferenceAPI
```

```python

# This uses https://huggingface.co/HuggingFaceH4/zephyr-7b-alpha
# downloaded (if first invocation) to the local Hugging Face model cache,
# and actually runs the model on your local machine's hardware
model_name = "microsoft/Phi-4-mini-instruct"

locally_run = HuggingFaceLLM(
    model_name=model_name,
    tokenizer_name=model_name,
    # I just put the querry in the right format below
    # query_wrapper_prompt=f"<|system|>{system_message}<|end|><|user|>{{query}}<|end|><|assistant|>",
    context_window=4096,
    max_new_tokens=256,
    device_map="auto"#,
    #trust_remote_code=True # **Crucial for many Microsoft models** according to gemini, threw an error for me
)
```

Underlying a completion with `HuggingFaceInferenceAPI` is Hugging Face's
[Text Generation task](https://huggingface.co/tasks/text-generation).

If you are modifying the LLM, you should also change the global tokenizer to match!

```python
system = "<|system|>"+"You are a helpful friend"+"<|end|>"
user = "<|user|>"+"How can I address my plantar fasciitis"+"<|end|>"
prompt = system + user + "<|assistant|>"
print(prompt)
response = locally_run.complete(prompt)
print(response)
```

```python
prompt = "<|system|>You are a helpful friend<|end|><|user|>How can I address my plantar fasciitis<|end|><|assistant|>"
response = locally_run.complete(prompt)
print(response)
```

```python
# @title Creating a simple chatbot
system = "" # @param {"type":"string"}
user  =  "How can I address my plantar fasciitus" # @param {"type":"string"}

prompt = "<|system|>" + system + "<|end|><|user|>" + user + "<|end|><|assistant|>"
response = locally_run.complete(prompt)
print(response)

```
